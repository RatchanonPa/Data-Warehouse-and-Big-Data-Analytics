**มีค่าผิดปกติ (outliers) และหรือข้อมูลรบกวน (noisy data) หรือไม่ในแต่ละตัวแปร?**
> Noisy: Data contains errors or outliers

```python
# prompt: คำนวณค่าผิดปกติ
# Calculate the IQR (Interquartile Range) for each numerical column
for col in titanic.select_dtypes(include=np.number):
  Q1 = titanic[col].quantile(0.25)
  Q3 = titanic[col].quantile(0.75)
  IQR = Q3 - Q1

  # Define the lower and upper bounds for outliers
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Identify outliers
  outliers = titanic[(titanic[col] < lower_bound) | (titanic[col] > upper_bound)]

  # Print the number of outliers in each column
  print(f"Outliers in {col}: {len(outliers)}")
```

```python
# Check for outliers using box plots
numeric_columns = titanic.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(12, 6))
titanic[numeric_columns].boxplot()
plt.title('Box Plot for Numeric Variables')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked

```python
# One-hot encode the 'embarked' column
embarked_encoded = pd.get_dummies(titanic['embarked'], prefix='embarked')
# Concatenate the encoded columns to the original DataFrame
titanic = pd.concat([titanic, embarked_encoded], axis=1)
# Drop the original 'embarked' column
titanic.drop('embarked', axis=1, inplace=True)
titanic
```

ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1

```python
# Map 'male' to 0 and 'female' to 1
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})
titanic
```

iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard)

```python
# Create FamilySize column
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']
# Create IsAlone column
titanic['IsAlone'] = np.where(titanic['FamilySize'] == 0, 1, 0)
titanic
```

iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# Create IsAlone column
titanic['IsAlone'] = np.where(titanic['FamilySize'] == 0, 1, 0)
titanic
```

i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the 'fare' and 'age' columns
titanic[['fare', 'age']] = scaler.fit_transform(titanic[['fare', 'age']])
titanic
```

ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1

```python
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)
titanic
```

iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
# Create AgeGroup column
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=[0, 12, 18, 50, 100], labels=['Child', 'Teenager', 'Adult', 'Senior'])
titanic
```

iii. (10 points) survived และ age

ใช้ **KDE Plot (Kernel Density Estimation Plot)** ซึ่งเป็นกราฟที่แสดงการกระจายตัวของอายุในกลุ่มผู้รอดชีวิต (`survived == 1`) และไม่รอดชีวิต (`survived == 0`) ได้


```python
# สร้าง KDE Plot สำหรับการกระจายของอายุตามการรอดชีวิต
plt.figure(figsize=(10,6))
sns.kdeplot(data=titanic[titanic['survived'] == 1]['age'], color='green', label='Survived', shade=True)
sns.kdeplot(data=titanic[titanic['survived'] == 0]['age'], color='red', label='Did not survive', shade=True)
# ปรับแต่งกราฟ
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Survival Status')
plt.grid(True)
plt.show()
```


> - เส้นกราฟสีเขียวแสดงการกระจายตัวของอายุในกลุ่มผู้รอดชีวิต ส่วนเส้นกราฟสีแดงแสดงในกลุ่มที่ไม่รอดชีวิต
- จากกราฟจะเห็นได้ว่าผู้โดยสารที่มีอายุน้อยกว่า (เช่น เด็ก) มีแนวโน้มรอดชีวิตสูงกว่าผู้โดยสารที่มีอายุสูงกว่า

* (10 points) สร้างตารางสรุปข้อมูลเพื่อหาคุณลักษณะของผู้โดยสารเรือไททานิค และศึกษาเกี่ยวกับการเสียชีวิตของผู้โดยสาร ทำการสรุปผลที่ได้จากตาราง

```python
pivot_table = pd.pivot_table(
    titanic,
    values='survived',  # The column to aggregate (Survived status)
    index=['sex', 'pclass', 'IsAlone'],  # Group by sex, pclass, and IsAlone
    aggfunc={'survived': ['count', 'sum', 'mean']}  # Aggregations: count (total passengers), sum (survived), mean (survival rate)
)

# Rename columns for clarity
pivot_table.columns = ['Total_Passengers', 'Total_Survived', 'Survival_Rate']

# Calculate Not Survived by subtracting Survived from Total Passengers
pivot_table['Not_Survived'] = pivot_table['Total_Passengers'] - pivot_table['Total_Survived']

# Reset index for a cleaner display
pivot_table = pivot_table.reset_index()
pivot_table
```
